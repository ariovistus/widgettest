```python
from waka import HelloWidget 
```

```python
%%javascript
require.undef("hello");
define('hello', ["jupyter-js-widgets"], function(widgets) {
    var HelloView = widgets.DOMWidgetView.extend({
        render: function() {
            
            //window.el = this.el;
            console.info(this.model.get('leds'));
            this.make_leds();
            this.model.on("change:leds", this.leds_changed, this);
            this.leds_changed();
            
        },
        leds_changed: function() {
            var ledsModel = this.model.get('leds');
            console.info(ledsModel);
            for(var i = 0; i < this.leds.length; i++) {
                var data = ledsModel[i];
                this.leds[i].style['background-color'] = "#" + this.bhex(data.red) + this.bhex(data.green) + this.bhex(data.blue);
            }
        },
        bhex: function(n) {
            var result = n.toString(16);
            if (result.length == 1) {
                result = "0" + result;
            }
            return result;
        },
        make_leds: function() {
            var nleds = this.model.get('nleds');
            var width = document.querySelector(".cell").offsetWidth; // haack!
            var leds_per_line = Math.floor(width * .8 / this.led_width());
            var nfullRows = Math.floor(nleds / leds_per_line);
            var remainderLeds = nleds % leds_per_line;
            console.info("nfull: ", nfullRows);
            console.info("r: ", remainderLeds);
            this.leds = [];
            for(var j = 0; j < nfullRows; j++) {
                this.make_led_row(leds_per_line, j);
            }
            console.info(this.make_led_row);
            this.make_led_row(remainderLeds, j);
        },
        make_led_row: function(leds_per_line, j) {
            var ledsInRow = [];
            console.info("badurk");
            for(var i = 0; i < leds_per_line; i++) {
                var led = this.make_led();
                ledsInRow.push(led);
            }
            
            var n = ledsInRow.length;
            for(var i = 0; i < n; i++) {
                var led;
                if(j % 2 == 0) {
                    led = ledsInRow.shift();
                }else{
                    led = ledsInRow.pop();
                }
                this.leds.push(led);
            }  
        },
        make_led: function() {
            var border = document.createElement("span");
            border.style.padding = this.border_padding + 'px';
            border.style['padding-top'] = '8px';
            border.style['padding-bottom'] = '8px';
            border.style['background-color'] = 'black';
            var led = document.createElement("span")
            led.style['background-color'] = 'green';
            led.style['padding'] = this.led_padding + 'px';
            led.style['padding-top'] = '3px';
            led.style['padding-bottom'] = '3px';
            border.appendChild(led);
            this.el.appendChild(border);
            return led;
        },
        make_newline: function() {
            var s = document.createElement("span");
            s.style.display='block';
            this.el.appendChild(s);
        },
        border_padding: 5,
        led_padding: 4,
        led_width: function() {
            return 2 * (this.border_padding + this.led_padding);
        }
    });
    
    return {
        HelloView: HelloView
    }
});
```
